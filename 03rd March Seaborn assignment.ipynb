```python
#Seaborn assignment
"""Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each 
plot."""

Ans: 
1. Line Plot: A line plot is used to visualize the relationship between two continuous variables. It can 
be used to compare trends over time or to show the correlation between two variables.

2. Bar Plot: A bar plot is used to compare the values of different categories or groups. It can be used 
to compare the means of different groups or to show the distribution of data within a group.

3. Box Plot: A box plot is used to visualize the distribution of a dataset. It can be used to identify 
outliers and compare the spread of different groups. groups or to show the distribution of data within a 
group.

4. Pair Plot: A pair plot is used to visualize the relationship between multiple variables. It can be used
to identify correlations between variables and to compare the strength of those correlations.

5. scatter Plot: A scatterplot is a type of data visualization that uses dots to represent values for two 
different variables. It is used to show the relationship between the two variables, with one variable 
plotted on each axis. Scatterplots are useful for identifying potential correlations in the data.

"""Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions. 
Note: timepoint, signal, event, and region are columns in the fmri dataset."""

Ans: 
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset('fmri')

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)


"""Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset."""

Ans: 
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset('titanic')

sns.boxplot(x = titanic.pclass, y = titanic.age)
sns.boxplot(x = titanic.pclass, y = titanic.fare)

"""Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset."""

Ans:
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

sns.histplot(diamonds.price,y = diamonds.cut)

"""Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset."""

Ans: 
import seaborn as sns
sns.pairplot(data=iris, hue="species")

"""Que 6: Use the "flights" dataset from seaborn to plot a heatmap."""
Ans: 
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights, annot=True, fmt="d")
```
